```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
df_diamonds= diamonds.copy()
```

# 2 violin plot

```python
sns.boxplot(x = 'color', y = 'price', data = df_diamonds, palette='Accent') # https://www.practicalpythonfordatascience.com/ap_seaborn_palette
plt.title("Seaborn diamonds dataset : Boxplot of Price by Color")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='color', y='price', data=df_diamonds, palette='CMRmap') # https://www.practicalpythonfordatascience.com/ap_seaborn_palette
plt.title('Seaborn diamonds dataset :Violinplot of Price by color')
plt.show()
```

```python
df_diamonds['cut'] = pd.factorize(df_diamonds['cut'])[0]
df_diamonds['cut']
```

```python
df_diamonds['cut'] = pd.factorize(df_diamonds['cut'])[0]
df_diamonds['color'] = pd.factorize(df_diamonds['color'])[0]
df_diamonds['clarity'] = pd.factorize(df_diamonds['clarity'])[0]


correlation_matrix = df_diamonds.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(data=correlation_matrix, cmap='RdBu', annot=True)
plt.title('Seaborn diamonds dataset : Correlation Heatmap')
plt.show()
```
